```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

from google.colab import files
import io
```

```python
unemployment_rate_data = pd.read_csv("https://raw.githubusercontent.com/n9hquan/QR103/main/data-ATqMs.csv")
unemployment_rate_data
```

```python
women_rate =  unemployment_rate_data.women
```

```python
men_rate =  unemployment_rate_data.men
```

```python
st.ttest_ind(women_rate, men_rate)
```

```python
st.mannwhitneyu(women_rate, men_rate)
```

```python
#Solution goes here
ci3 = st.norm.interval(0.95, loc=women_rate.mean(),scale=women_rate.sem())
ci4 = st.norm.interval(0.95, loc=men_rate.mean(),scale=men_rate.sem())
plt.figure(figsize=(6,6))
plt.bar(1, women_rate.mean(), width = 0.1, capsize=10, yerr = (ci3[1]-ci3[0])/2, color="lightpink")
plt.bar(1.15, men_rate.mean(), width = 0.1, capsize=10, yerr = (ci4[1]-ci4[0])/2, color="darkblue")
plt.xticks([1,1.15],['Women unemployment rate','Men unemployment rate'])
plt.show()
```

```python
st.shapiro(women_rate)
```

```python
labour_rate_data = pd.read_csv("https://raw.githubusercontent.com/n9hquan/QR103/main/data-SfSwY.csv")
labour_rate_data
```

```python
men_labour = labour_rate_data.men
```

```python
women_labour = labour_rate_data.women
```

```python
st.shapiro(men_labour)
```

```python
st.ttest_ind(women_labour, men_labour)
```

```python
ci5 = st.norm.interval(0.95, loc=women_labour.mean(),scale=women_labour.sem())
ci6 = st.norm.interval(0.95, loc=men_labour.mean(),scale=men_labour.sem())
plt.figure(figsize=(6,6))
plt.bar(1, women_labour.mean(), width = 0.1, capsize=10, yerr = (ci5[1]-ci5[0])/2, color = "lightpink")
plt.bar(1.25, men_labour.mean(), width = 0.1, capsize=10, yerr = (ci6[1]-ci6[0])/2, color = "darkblue")
plt.xticks([1,1.25],['Women labour force participation rate ',' Men labour force participation rate'])
plt.show()
```
