```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import random
from scipy.stats import norm
import scipy.stats as st
import seaborn as sns
from numpy import random
```

```python
values = np.random.uniform(0, 1, 1000)
count, bins, ignored = plt.hist(values, 15, density=True)
plt.plot(bins, np.ones_like(bins),color='r')
plt.title('Uniform Distribution')
plt.ylabel('Density')
plt.xlabel('Values')
plt.show()
```

```python
list_of_means = []
for i in range(1000):
  sample = random.sample(list(values), k=20)
  mean = sum(sample) / len(sample)
  list_of_means.append(mean)
plt.hist(list_of_means, bins=25, density=True, color='b')
mu1 = sum(list_of_means) / len(list_of_means)
sigma1 = np.std(list_of_means)
x1 = np.linspace(min(list_of_means),max(list_of_means), 100)
plt.plot(x1, st.norm.pdf(x1, mu1, sigma1), color = 'r');
plt.hist(list_of_means, bins = np.arange(min(list_of_means),max(list_of_means), 3), density = True);
plt.show()
```

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

#generate exponential distribution with sample size 10000
exponential_values = expon.rvs(scale=100, size=20)

#create plot of exponential distribution
plt.hist(exponential_values, density=True, edgecolor='black')
```

```python
list_of_means_2 = []
for i in range(500):
  exponential_values = expon.rvs(scale=100, size=20)
  mean_2 = exponential_values.mean()
  list_of_means_2.append(mean_2)
plt.hist(list_of_means_2, bins=25, density=True, color='b')
mu2 = sum(list_of_means_2)/len(list_of_means_2)
sigma2 = np.std(list_of_means_2)
x2 = np.linspace(min(list_of_means_2),max(list_of_means_2), 100)
plt.plot(x2, st.norm.pdf(x2, mu2, sigma2), color = 'r');
plt.hist(list_of_means_2, bins = np.arange(min(list_of_means_2),min(list_of_means_2), 3), density = True);
plt.show()
```
